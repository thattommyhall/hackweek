```python
def remote_execute(address, fn_cid, arg_cid, verbose=False):
    uri = f"{address}/call/{fn_cid}/on/{arg_cid}"
    if verbose:
        print(f"hitting {uri}")
    r = requests.get(uri)
    r.raise_for_status()
    result_cid = r.json()['result']
    if verbose:
        print(f"Result at {result_cid}")
    return ipfs_cat(result_cid)
```

```python
client = ipfshttpclient.connect()

count = 0

def get_blocks(cid, depth=0): 
    result = []
    objects = client.ls(cid)['Objects']
    links = objects[0]['Links']
    if len(links) == 0:
        return [objects[0]['Hash']]
    for link in links:
        result.extend(get_blocks(link['Hash']))
    return result 
    
```
